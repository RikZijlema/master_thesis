```python
import nltk
import time
import pandas as pd
nltk.download("punkt")
pd.options.display.max_rows = 10000
pd.options.display.max_columns = 10000
pd.options.display.max_colwidth = None

train = pd.read_csv("/content/drive/MyDrive/Master Thesis/train_lilah.csv")
dev = pd.read_csv("/content/drive/MyDrive/Master Thesis/dev_lilah.csv")
test = pd.read_csv("/content/drive/MyDrive/Master Thesis/test_lilah.csv")

lilah = pd.read_csv("/content/drive/MyDrive/Master Thesis/LiLaH-HR-NL-SL.tsv", sep="\t")
#Remove the languages that are not needed
lilah = lilah.drop(["HR","SL"], axis=1)
#Drop rows without any word in them
lilah = lilah[lilah["NL"] != "_"]
lilah = lilah[lilah["NL"] != "NO TRANSLATION"]

def extract_emotions(article, emotion):
    article = article.lower()
    article_word_list = nltk.word_tokenize(article)
    word_dict = {}
    for word, score in zip(lilah["NL"], lilah[emotion]):
        lilah_word_list = word.split(",")
        intersect_list = tuple(set(article_word_list) & set(lilah_word_list))
        for word in intersect_list:
            word_dict[word] = score
    if bool(word_dict) == True:
         return sum(word_dict.values()) / len(word_dict)
    if bool(word_dict) == False:
         return 0

train["pos_count"] = train["article"].apply(extract_emotions, emotion="Positive")
dev["pos_count"] = dev["article"].apply(extract_emotions, emotion="Positive")
test["pos_count"] = test["article"].apply(extract_emotions, emotion="Positive")

train["neg_count"] = train["article"].apply(extract_emotions, emotion="Negative")
dev["neg_count"] = dev["article"].apply(extract_emotions, emotion="Negative")
test["neg_count"] = test["article"].apply(extract_emotions, emotion="Negative")

train.to_csv("/content/drive/MyDrive/Master Thesis/train_lilah.csv", index=False)
dev.to_csv("/content/drive/MyDrive/Master Thesis/dev_lilah.csv", index=False)
test.to_csv("/content/drive/MyDrive/Master Thesis/test_lilah.csv", index=False)
```
